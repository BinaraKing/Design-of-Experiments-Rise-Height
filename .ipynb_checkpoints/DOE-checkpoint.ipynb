# **Design of Experiment - Rise of Muffin Height**

```python
%reset
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import stats
from scipy.stats import norm
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```



based on our data we can above we can code our units to be -1 or +1 fror the predictor variables

```python
## creating data frame 
temp=pd.Series([350,350,350,350,350,350,350,350,250,250,250,250,250,250,250,250],name='temperature')
soda=pd.Series([1,1,2,2,1,1,2,2,1,1,2,2,1,1,2,2], name='baking soda')
time=pd.Series([15,15,15,15,25,25,25,25,15,15,15,15,25,25,25,25], name='time')
height=pd.Series([48,49,53,52,48,49,54,54,36,37,42,43,39,40,44,43], name='muffin height')

DF= pd.concat([temp,soda,time,height], axis='columns')

# randomize the order of the experiments
DF = DF.sample(frac=1)

print(DF)
```

```python
## normalize data

Temp = DF['temperature'].replace([250,350],[-1,1])
Soda = DF['baking soda'].replace([1,2],[-1,1])
Time = DF['time'].replace([15,25],[-1,1])
Height = DF['muffin height']

data= pd.concat([Temp,Soda,Time,Height], axis='columns')
print(data)
```

```python
## build preliminary OLS Model
model1 = ols("Height~Temp*Soda*Time",data=data)
results1 = model1.fit()
print(results1.summary2())

#QQ Plot of Residuals for preliminary model
DOEQQ1 = sm.qqplot(results1.resid,line='s')
plt.title('QQ Plot of Residuals of preliminary model')
plt.show()
DOEQQ1 = DOEQQ1.get_figure()
DOEQQ1.savefig("DOEQQ1.png")
```

As you can see, all of the double factored terms are insignificant an can therefore be romoved from our model. however, are predictor slopes are significant and the tripple factor terms is also significant and can therefore be included in our model 


```python
## refined model
model2 = ols("Height~Temp*Soda*Time-Temp*Soda-Soda*Time+Temp+Soda+Time",data=data)
results2 = model2.fit()
print(results2.summary2())

#QQ Plot of Residuals for new model
DOEQQ2 = sm.qqplot(results2.resid,line='s')
plt.title('QQ Plot of Residuals of new model')
plt.show()
DOEQQ2 = DOEQQ2.get_figure()
DOEQQ2.savefig("DOEQQ2.png")
```

## Temp Vs Baking Soda Model

```python
## Create OLS Model

model3 = ols("Height~Temp*Soda",data = data)
results3 = model3.fit()

#Set up Factors
df = data
x = df['temperature']
y= df['baking soda']
z = df['muffin height']
x1,x2,x3 = model3.exog_names[1:]
x_surf = np.arange(-1.1,1.1,0.01)
y_surf = np.arange(-1.1,1.1,0.01)
xS,yS = np.meshgrid(x_surf,y_surf)
exog = pd.DataFrame({x1: xS.ravel(), x2:yS.ravel()})
zS = results3.predict(exog=exog).values.reshape(xS.shape)

## interaction plots

#Initial Concentration vs Temperature Interaction Plot
fig, ax = plt.subplots(1,2,sharey=True)
sns.lineplot(x="temperature",y="muffin height",hue="baking soda",data=data,ax=ax[0])
sns.lineplot(x="baking soda",y="muffin height",hue="temperature",data=data,ax=ax[1])
fig.suptitle("Interaction Plots of muffin height (mm) with temperature and baking soda)")
plt.show()

## contour plot

fig, ax = plt.subplots(1, 1)
contour3=ax.contourf(xS,yS,zS,cmap="coolwarm")
fig.colorbar(contour3,shrink=0.5,aspect=5)
plt.xlabel("Temperature")
plt.ylabel("Baking Soda")
plt.title("Contour Plot of Muffin Height (mm) with temperature and baking soda")
plt.show()

```
